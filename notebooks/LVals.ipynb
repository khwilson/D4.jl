```julia
function l2(D::Int64, N::Int64)
    total::Float64 = 0.0
    for i in 1:N
        total += D4.kron(D, i) / i / i
    end
    total
end

function l1(D::Int64, N::Int64)
    total::Float64 = 0.0
    for i in 1:N
        total += D4.kron(D, i) / i
    end
    total
end
```

```julia
max_D = 10000000
sqfreeIntegers = ones(Bool, max_D)
sqfreeIntegers[1] = false
for D in 2:length(sqfreeIntegers)
    if sqfreeIntegers[D]
        Dsquared = D * D
        val = Dsquared
        while val < length(sqfreeIntegers)
            sqfreeIntegers[val] = false
            val += Dsquared
        end
    end
end
```

```julia
numTerms = 1000000
open("l2_values", "w") do f
    for d in 1:length(sqfreeIntegers)
        if d % 100 == 1
            @printf "At %d\n" d
        end
        if d == 4
            print(f, "$(-d),$(l2(-d, numTerms))\n")
            continue
        end
        if d % 4 == 0
            D = div(d, 4)
            if sqfreeIntegers[D] && ((D % 4 == 2) || (D % 4 == 1))
                print(f, "$(-d),$(l2(-d, numTerms))\n")
            end
        end
        if d % 4 == 3 && sqfreeIntegers[d]
            print(f, "$(-d),$(l2(-d, numTerms))\n")
        end
    end
end
```

```julia
function compute(sqfreeIntegers::Array{Bool, 1}, numTerms1::Int64, numTerms2::Int64)

    total = 0.0

    for d in 1:length(sqfreeIntegers)
        if d % 100 == 1
            @printf "Current total at d=%d: %.12f\n" d total
        end
        if !sqfreeIntegers[d]
            continue
        end

        if d % 4 == 1
            total += l1(d, numTerms1) / l2(d, numTerms2) / d
            total += 0.125 * l1(-4 * d, numTerms1) / l2(-4 * d, numTerms2) / d
        elseif d % 4 == 2
            d = 4 * d
            total += l1(d, numTerms1) / l2(d, numTerms2) / d
            total += 0.5 * l1(-d, numTerms1) / l2(-d, numTerms2) / d
        elseif d % 4 == 3
            total += l1(-d, numTerms1) / l2(-d, numTerms2) / d
            d = 4 * d
            total += 0.5 * l1(d, numTerms1) / l2(d, numTerms2) / d
        end
    end
    total / Ï€^2
end
```
